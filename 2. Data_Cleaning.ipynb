```python
import os
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

save_loc = 'C:\\Users\\calvi\\OneDrive\\Desktop\\database\\curry_prop'
os.chdir(save_loc)
```

# Reading Data

```python
df_10_17 = pd.read_csv('game_logs2010-2017v2.csv', index_col=0)
df_18_22 = pd.read_csv('game_logs18-23.csv', index_col=0)
df_22_23 = pd.read_csv('game_logs22-23v3.csv', index_col=0)
df = pd.concat([df_10_17, df_18_22, df_22_23], axis=0)
df.reset_index(inplace=True, drop=True)
```

# Data Cleaning/Modifying Columns

```python
name_map = df_player_info.set_index('index').to_dict()['Player']
df['Player'] = df['Index'].map(name_map)

pos_map = df_player_info.set_index('index').to_dict()['Pos']
df['Pos'] = df['Index'].map(pos_map)

ht_map = df_player_info.set_index('index').to_dict()['Ht']
df['Ht'] = df['Index'].map(ht_map)

wt_map = df_player_info.set_index('index').to_dict()['Wt']
df['Wt'] = df['Index'].map(wt_map)

df = df[df.columns[-5:].tolist() + df.columns[:-5].tolist()]
```

```python
# The html scraper picked up headings, so I drop those rows 
df.drop(index=df.loc[df['Date'] == 'Date'].index, inplace=True)
df['Date'] = df['Date'].astype('str')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Deleting games where a player didn't play
df = df.drop(index=df.loc[df['PTS'] == 'Inactive'].index)
df = df.drop(df.loc[df['PTS'] == 'Did Not Dress'].index)
df = df.drop(df.loc[df['PTS'] == 'Did Not Play'].index)
df = df.drop(df.loc[df['PTS'] == 'Not With Team'].index)
df = df.drop(df.loc[df['PTS'] == 'Player Suspended'].index)
```

```python
# Creating Home/Away Column
df = df.rename(columns = {'Unnamed: 5' : 'Home/Away'})
df['Home/Away'] = np.where(df['Home/Away'] == '@', 'Away', 'Home')
```

```python
# FG%, 3P% and FT% are null if there are no attempts, so I just set it to 0 (need to think about better way to handle this)
df.loc[df['FG%'].isna(), 'FG%'] = 0.0
df.loc[df['3P%'].isna(), '3P%'] = 0.0
df.loc[df['FT%'].isna(), 'FT%'] = 0.0
```

```python
# Converting MP into whole number
df.insert(10, 'MP_num', df['MP'].apply(lambda x : float(x.split(':')[0]) + float(x.split(':')[1])/60))
```

```python
df.insert(4, 'Age_whole', df['Age'].apply(lambda x : int(x.split('-')[0])))
```

```python
df['GS'] = df['GS'].astype('int')
df['FG'] = df['FG'].astype('int')
df['FGA'] = df['FGA'].astype('int')
df['FG%'] = df['FG%'].astype('float')
df['3P'] = df['3P'].astype('int')
df['3PA'] = df['3PA'].astype('int')
df['3P%'] = df['3P%'].astype('float')
df['FT'] = df['FT'].astype('int')
df['FTA'] = df['FTA'].astype('int')
df['FT%'] = df['FT%'].astype('float')
df['ORB'] = df['ORB'].astype('int')
df['DRB'] = df['DRB'].astype('int')
df['TRB'] = df['TRB'].astype('int')
df['STL'] = df['STL'].astype('int')
df['BLK'] = df['BLK'].astype('int')
df['TOV'] = df['TOV'].astype('int')
df['PF'] = df['PF'].astype('int')
df['PTS'] = df['PTS'].astype('int')
```

```python
# Adding in season
df.insert(3, 'Season', 'year')
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2009-10-27')) & (df['Date'] <= pd.to_datetime('2010-4-14')), '09-10', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2010-10-26')) & (df['Date'] <= pd.to_datetime('2011-4-13')), '10-11', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2011-12-25')) & (df['Date'] <= pd.to_datetime('2012-4-26')), '11-12', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2012-10-30')) & (df['Date'] <= pd.to_datetime('2013-4-17')), '12-13', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2013-10-29')) & (df['Date'] <= pd.to_datetime('2014-4-16')), '13-14', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2014-10-28')) & (df['Date'] <= pd.to_datetime('2015-4-15')), '14-15', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2015-10-27')) & (df['Date'] <= pd.to_datetime('2016-4-13')), '15-16', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2016-10-25')) & (df['Date'] <= pd.to_datetime('2017-4-13')), '16-17', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2017-10-17')) & (df['Date'] <= pd.to_datetime('2018-4-11')), '17-18', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2018-10-16')) & (df['Date'] <= pd.to_datetime('2019-4-10')), '18-19', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2019-10-22')) & (df['Date'] <= pd.to_datetime('2020-8-14')), '19-20', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2020-12-22')) & (df['Date'] <= pd.to_datetime('2021-5-16')), '20-21', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2021-10-19')) & (df['Date'] <= pd.to_datetime('2022-4-10')), '21-22', df['Season'])
df['Season'] = np.where((df['Date'] >= pd.to_datetime('2022-10-18')) & (df['Date'] <= pd.to_datetime('2023-4-9')), '22-23', df['Season'])
```
