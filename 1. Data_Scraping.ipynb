# 1. Scraping Game Logs
I am scraping game logs from basketball-reference.com which have basic box score stats such as points, FGA, assists, etc.

## 1.1 Grabbing Player Info
Here I am grabbing a list of all NBA players that are listed on basketball-reference.com

```python
alphabet = list(string.ascii_lowercase)
list_of_dfs = []

for letter in alphabet:
    print(letter)
    try:
        url = "https://www.basketball-reference.com/players/{}".format(letter)
        # this is the HTML from the given URL
        html = urlopen(url)
        soup = BeautifulSoup(html)

        headers = [th.getText() for th in soup.findAll('tr')[0].findAll('th')]
        headers = headers

        rows = soup.findAll('tr')

        player_names = [[td.getText() for td in rows[i].findAll('th')]
                    for i in range(len(rows))]



        names = pd.DataFrame(player_names, columns = headers)
        names.head(10)

        player_stats = [[td.getText() for td in rows[i].findAll('td')]
                    for i in range(len(rows))]


        stats = pd.DataFrame(player_stats, columns = headers[1:])
        stats['Player'] = names['Player']
        indexes = []
        for i in range(0, len(rows)):
            index = rows[i].find('th').get('data-append-csv')
            indexes.append(index)

        stats['index'] = indexes

        list_of_dfs.append(stats)
        time.sleep(20)
    except:
        continue
```

```python
new_df = pd.concat(list_of_dfs)
new_df = new_df[1:]
new_df = new_df.loc[new_df['Player'] != 'Player']
new_df
```

```python
# Saving player info
new_df.to_csv('player_info.csv')
```

## 1.2 Scraping Game Logs
After grabbing player info, I use a for loop to go through a list of players and scrape all of their game log data in a list of seasons.

```python
# Grabbing names/id of players that play until at least 2008
players_from_2008 = list(new_df.loc[new_df['To'] >= '2008']['index'])
```

```python

total_player_dfs = []

for player in players_from_2008:
    letter = player[0]
    years = [number for number in np.arange(2010, 2018)] # 2010/11 season through 2017/18 season
    for year in years:
        try:
            time.sleep(2)
            df = pd.read_html('https://www.basketball-reference.com/players/{}/{}/gamelog/{}'.format(letter, player, year))[7]
            df['Index'] = player
            players_name = new_df[new_df['index'] == player]['Player']
            df['Player'] = players_name
            total_player_dfs.append(df)
        except:
            continue
```

```python
# Joining all the separate game logs together into a single dataframe
new_totals = pd.concat(total_player_dfs)
```
